# Matplotlib example workflow

```python
#0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup data : Create figure and then create axis. Axis object means a subplot/s. (10,10) is (width,height)
fig, ax = plt.subplots(figsize=(10,10))

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")

#5. Save and show (we save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two options for Subplots

```python
# Subplots option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5));

# Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use ? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advances, use the OO method

```python
heart_disease.head()
```

```python
# Retrive age column with value > 50 usinf Boolean indexing
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age',y='chol', c='target');
```

```python
## OO Method mixed with pyplot method
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

#ax.set_xlim([45,100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], y=over_50['chol'], c = over_50['target']);

# Customize the plot
ax.set(title="Heart Disease and Cholestrol levels", xlabel="Age", ylabel="Cholesterol");

# Add a legend
# *scatter.legend_elements()= this * says that grab the values from scatter and unpack the legend elements which queries the 
# 'c' parameter (color) to look back into the dataframe and look what values are in this target column
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

```python
over_50["chol"].mean()
```

```python
over_50.head()
```

```python
# Subplot of chol,age,thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")

# Add legent to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line to ax0
ax0.axhline(over_50["chol"].mean(), linestyle='--');


## now move to ax1 subplot

# Add data to ax0
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize ax0
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

# Add legent to ax0
ax1.legend(*scatter.legend_elements(), title="Target")

# Add mean line to ax0
ax1.axhline(over_50["thalach"].mean(), linestyle='--')


# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50['age'], 
                     y=over_50['chol'], 
                     c = over_50['target'], 
                     cmap='winter'); # cmap changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholestrol levels", xlabel="Age", ylabel="Cholesterol");

# Add a legend
# *scatter.legend_elements()= this * says that grab the values from scatter and unpack the legend elements which queries the 
# 'c' parameter (color) to look back into the dataframe and look what values are in this target column
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--');
```

### This plot shows some information about the heart disease dataset...

```python
# Customizing the y and x axis limitations

# Subplot of chol,age,thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex=True);

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")

# Change the x-axis limits..removes those edges, make it clean
ax0.set_xlim([50,80])

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line to ax0
ax0.axhline(over_50["chol"].mean(), linestyle='--');


## now move to ax1 subplot

# Add data to ax0
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")

# Customize ax0
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")

# Change ax1 x and y axis limits
ax1.set_xlim([50,80]) # not neccessary to do for xlim as both share common x-axis
ax1.set_ylim([60,200])

# Add legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")

# Add mean line to ax0
ax1.axhline(over_50["thalach"].mean(), linestyle='--')


# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
